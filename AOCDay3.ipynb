# Advent of Code Day 3
#### Problem in full can be found here: https://adventofcode.com/2023/day/3

```python
# part 1
# Input file consists of numbers and symbols. Add all the numbers that have a symbol, thats not a period, vertically,
# horizontally, or diagonally
input_file_path = "inputday3.txt"
with open(input_file_path, 'r') as file:
    # Read lines from the file and create a 2D array
    lines = [list(line.strip()) for line in file]
sum = 0
num = ""
symbolExists = False
index = -1
for line in lines:
    index += 1
    if num != "" and symbolExists:
        sum += int(num)
    num = ""
    symbolExists = False
    for i in range(0, len(line)):
        if line[i].isnumeric():
            num += line[i]
            # Checking left side for symbol
            if i > 0:
                if not (line[i-1].isnumeric() or line[i-1] == '.'):
                    symbolExists = True
                if index > 0:
                    if not (lines[index-1][i-1].isnumeric() or lines[index-1][i-1] == '.'):
                        symbolExists = True
                if index < len(lines)-1:
                    if not (lines[index+1][i-1].isnumeric() or lines[index+1][i-1] == '.'):
                        symbolExists = True
            # Checking right side for symbol
            if i < len(line)-1:
                if not (line[i+1].isnumeric() or line[i+1] == '.'):
                    symbolExists = True
                if index > 0:
                    if not (lines[index-1][i+1].isnumeric() or lines[index-1][i+1] == '.'):
                        symbolExists = True
                if index < len(lines)-1:
                    if not (lines[index+1][i+1].isnumeric() or lines[index+1][i+1] == '.'):
                        symbolExists = True
            # Checking up and down for symbol
            if index > 0:
                if not (lines[index-1][i].isnumeric() or lines[index-1][i] == '.'):
                    symbolExists = True
            if index < len(lines)-1:
                if not (lines[index+1][i].isnumeric() or lines[index+1][i] == '.'):
                    symbolExists = True
        # If the character is a symbol (doesn't matter if its a period)        
        else:
            if num != "" and symbolExists:
                sum += int(num)
                num = ""
                symbolExists = False
            elif num != "":
                num = ""
file.close()
print(sum)
```

```python
# part 2
# Input file consists of numbers and symbols. Add the multiplication where the symbol is * and has exactly two numbers
# associated with it horizontally, vertically, or diagonally
input_file_path = "inputday3.txt"
with open(input_file_path, 'r') as file:
    # Read lines from the file and create a 2D array
    lines = [list(line.strip()) for line in file]
sum = 0
index = -1
nums = set()
num = ""
for line in lines:
    index += 1
    for i in range(0, len(line)):
        if line[i] == "*":
            # Check the top
            if index > 0 and i > 0:
                if lines[index-1][i-1].isnumeric():
                    start = i-1
                    while start > 0 and lines[index-1][start-1].isnumeric():
                        start -= 1
                    while lines[index-1][start].isnumeric():
                        num += lines[index-1][start]
                        start += 1
                    nums.add(int(num))
                    num = ""
            if index > 0 and i < len(line)-1:
                if lines[index-1][i+1].isnumeric():
                    start = i+1
                    while start > 0 and lines[index-1][start-1].isnumeric():
                        start -= 1
                    while start < len(line) and lines[index-1][start].isnumeric():
                        num += lines[index-1][start]
                        start += 1
                    nums.add(int(num))
                    num = ""
            if index > 0 and i > 0 and i < len(line)-1:
                if lines[index-1][i].isnumeric() and not lines[index-1][i+1].isnumeric() and not lines[index-1][i-1].isnumeric():
                    num = lines[index-1][i]
                    nums.add(int(num))
                    num = ""
            # Check the bottom
            if index < len(lines)-1 and i > 0:
                if lines[index+1][i-1].isnumeric():
                    start = i-1
                    while start > 0 and lines[index+1][start-1].isnumeric():
                        start -= 1
                    while lines[index+1][start].isnumeric():
                        num += lines[index+1][start]
                        start += 1
                    nums.add(int(num))
                    num = ""
            if index < len(lines)-1 and i < len(line)-1:
                if lines[index+1][i+1].isnumeric():
                    start = i+1
                    while start > 0 and lines[index+1][start-1].isnumeric():
                        start -= 1
                    while start < len(line) and lines[index+1][start].isnumeric():
                        num += lines[index+1][start]
                        start += 1
                    nums.add(int(num))
                    num = ""
            if index < len(lines) and i > 0 and i < len(line)-1:
                if lines[index+1][i].isnumeric() and not lines[index+1][i+1].isnumeric() and not lines[index+1][i-1].isnumeric():
                    num = lines[index+1][i]
                    nums.add(int(num))
                    num = ""
            # Check left and right
            if i > 0:
                if lines[index][i-1].isnumeric():
                    start = i-1
                    while start > 0 and lines[index][start-1].isnumeric():
                        start -= 1
                    while lines[index][start].isnumeric():
                        num += lines[index][start]
                        start += 1
                    nums.add(int(num))
                    num = ""
            if i < len(line)-1:
                if lines[index][i+1].isnumeric():
                    start = i+1
                    while lines[index][start].isnumeric():
                        num += lines[index][start]
                        start += 1
                    nums.add(int(num))
                    num = ""
            # If the criteria fits
            if len(nums) == 2:
                result = 1
                for number in nums:
                    result *= number
                sum += result
            nums = set()
file.close()
print(sum)
```
